# Fundamentos de Python para Datos

## Obtener la data via API

```python
import requests


url = "https://api-colombia.com/api/v1/Department"

# Obtener los datos
response = requests.get(url)


# Validar errores
if response.status_code != 200:
    print("Error obteniendo los datos:", response)

# Parsear datos a JSON
data = response.json()

```

```python
data
```

## Cargar data con Pandas

```python
import pandas as pd

colombia_df = pd.DataFrame(data)
```

```python
# obtener primeras filas
colombia_df.head()
```

```python
# tamaño del dataset
colombia_df.shape
```

```python
#obtener columnas
colombia_df.columns
```

```python
# información estadistica
colombia_df.describe()
```

```python
# información de datos nulos
colombia_df.info()
```

```python
# filtrar datos
max_population = colombia_df.loc[colombia_df["population"].idxmax()]
print(max_population)
```

```python
min_population = colombia_df.loc[colombia_df["population"].idxmin()]
print(min_population)
```

```python
# obtener columnas con datos nulos
colombia_df.isna().any()
```

```python
# borrar columnas
df_without_country = colombia_df.drop(columns='country', axis=1)
```

```python
import matplotlib.pyplot as plt


categories = colombia_df["name"]
values = colombia_df["surface"]

plt.barh(categories, values, color='skyblue')
plt.xlabel('Superficie')
plt.ylabel('Departamentos')
plt.title('Superficie por Departamento en Colombia')
plt.show()
```

```python
df = colombia_df.sort_values(by=['surface'])

categories = df["name"]
values = df["surface"]

plt.barh(categories, values, color='skyblue')
plt.xlabel('Superficie')
plt.ylabel('Departamentos')
plt.title('Superficie por Departamento en Colombia')
plt.show()
```
